```python
import pandas

movie_db = pandas.read_csv('./moviedb.csv')
movie_db = movie_db.fillna('')
movie_db['duration'] = movie_db['duration'].str.replace(',', '')
movie_db['duration'] = movie_db['duration'].str.replace(' ', '')
movie_db['duration'] = movie_db['duration'].str.replace('m', '')
movie_db['duration'] = movie_db['duration'].str.replace('h', ':')

movie_db
```

```python
movie_db['years'] = movie_db['date'].str.extract(r', (\d+)')
movie_db
```

1) Quel est le film le plus long ?

```python
column = movie_db['duration']
max_value = column.max()
print(max_value)
index = movie_db.index[movie_db['duration'] == max_value]
res = movie_db['title'][index[0]]
print(res)
```

2) Quels sont les 5 films les mieux notés ?

```python
highest_rated = movie_db['note'].nlargest(n=5)
indexes = highest_rated.index
res_highest = movie_db.iloc[indexes]
res_highest
```

4) Quels sont les 3 meilleurs films d’horreur ? Dramatique ? Comique ?

```python
horror_indexes = movie_db.index[movie_db['genre'].str.contains('Horror')].tolist()
drama_indexes = movie_db.index[movie_db['genre'].str.contains('Drama')].tolist()
# drama_indexes = movie_db.index[movie_db['genre'].str.contains('Drama')].tolist()
a = movie_db.iloc[horror_indexes]
res_highest_horror = a['note'].nlargest(n=3)
b = res_highest_horror.index
movie_db.iloc[b]
```

5) Parmi les 100 films les mieux notés, quel pourcentage sont américains ? Français ?

```python
top_100_highest_rated = movie_db['note'].nlargest(n=100)
top_100_indexes = top_100_highest_rated.index
res_top_100_highest = movie_db.iloc[top_100_indexes]

best_100_us_movies = res_top_100_highest.index[res_top_100_highest['original_country'].str.contains('United States')].tolist()
best_100_fr_movies = res_top_100_highest.index[res_top_100_highest['original_country'].str.contains('France')].tolist()
print(f'Pourcentage de films Français :  {len(best_100_us_movies)} %')
print(f'Pourcentage de films Français :  {len(best_100_fr_movies)} %')
```

6) Quel est la durée moyenne d’un film en fonction du genre ?

```python
Sci_fi_indexes = movie_db.index[movie_db['genre'].str.contains('Sci-Fi')].tolist()
Sci_fi_movies = movie_db.iloc[Sci_fi_indexes]
# Sci_fi_movies
duration_mean = Sci_fi_movies['duration'].mean()
duration_mean
```
